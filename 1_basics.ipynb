## 2. Vector Addition and Subtraction

```python
def plot2D_vector(v,axis=[-10,10,-10,10]):
    plt.plot([0,v[0]],[0,v[1]])
    plt.axis(axis)
    plt.grid()
```

```python
#a+b = b+a; commutative
v1 = np.array([2,4])
plot2D_vector(v1)
```

```python
plot2D_vector((v1*-1))
```

```python
"""
Create a 2D vector v, 10 scalars that are drawn at random from normal distribution 
and plot all 10 scalar vector multiplication on top of each.
What did you observe?
"""
```

```python
v = np.array([2,3])

plt.plot([0,v[0]],[0,v[1]])

for i in range(10):

    s = np.random.randn()
    sv = s*v
    plt.plot([0,sv[0]],[0,sv[1]],alpha=0.1)
plt.grid()
plt.axis([-10,10,-10,10])
```
